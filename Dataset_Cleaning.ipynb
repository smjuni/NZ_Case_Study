```python
#convert order_date to datetime format
sales_df["order_date"]=pd.to_datetime(sales_df["order_date"], yearfirst=True)
sales_df.dtypes
```

```python
#How many NANs are in the dataset
sales_df.isnull().sum().sum()
```

```python
#Find the nans - Are they in multiple columns
sku_nans=sales_df["skus_count"].isnull().values.sum()
# sales_df["customer_id"].isnull().values.sum()
# sales_df["order_id"].isnull().values.sum()
order_item_id_nans = sales_df["order_item_id"].isnull().values.sum()
# sales_df["new_or_repeat"].isnull().values.sum()
print(f"Incomplete SKU count: {sku_nans}.")
print(f"Missing Order items ID: {order_item_id_nans}.")
print(f"Total NANs = {sku_nans + order_item_id_nans}.")
```

```python
#Check for patterns - All the nans are the same product_category and item_sku
sku_df=sales_df.loc[sales_df["skus_count"].isnull()]
print(sku_df.nunique())
sku_df
```

```python
#for tableau ease, create 2 separate columns for new/repeat customers (only because binary)
sales_df_dummies=pd.get_dummies(data=sales_df, columns=["new_or_repeat_by_order"], prefix="cust_type")
sales_df_dummies
```

```python
sales_df_dummies["new_or_repeat_by_order"] = sales_df["new_or_repeat_by_order"]
sales_df_dummies
```

```python
repeat_count=sales_df_dummies.groupby(sales_df_dummies["new_or_repeat_by_order"]=="Repeat").sum()
repeat_count
```
