```python
#import dependencies
import json
import pandas as pd
```

```python
#file path
file_dir="datasets_for_test/Users/keichunlaw/noissue/ingestion/zd/nps_surveys.json"
```

```python
#open json file
with open(f'{file_dir}', mode="r") as file:
    survey_raw = json.load(file)
```

```python
#view json file in readable format
survey_formatted = json.dumps(survey_raw, indent=2)

print(survey_formatted)
```

```python
#superficially investigate the file
all_rows_df =pd.read_json(file_dir)

print(len(all_rows_df))
all_rows_df
```

```python
#flatten the json file to extract all data
survey_data = pd.json_normalize(data=survey_raw['all_rows'], record_path='data')
survey_data
```

```python
#check for null values
survey_data["rating"].isnull().sum()
```
